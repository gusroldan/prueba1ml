Realizamos los importes necesarios para poder realizar los analisis


```python

import pandas as pd  # Biblioteca principal para el manejo de datos en DataFrames
import numpy as np  # Biblioteca para trabajar con arrays y operaciones matemáticas

import matplotlib.pyplot as plt  # Biblioteca para crear gráficos y visualizaciones
from matplotlib import style  # Estilo de gráficos de matplotlib
import seaborn as sns  # Biblioteca para mejorar la visualización con gráficos estadísticos

from scipy.stats import pearsonr  # Función para calcular la correlación de Pearson
from sklearn.linear_model import LinearRegression  # Modelo de regresión lineal de scikit-learn
from sklearn.model_selection import train_test_split  # Función para dividir los datos en conjunto de entrenamiento y prueba
from sklearn.metrics import r2_score  # Métrica para calcular el R^2 (coeficiente de determinación)
from sklearn.metrics import mean_squared_error  # Métrica para calcular el error cuadrático medio (MSE)

# Configuración de parámetros globales de visualización

plt.rcParams['image.cmap'] = "bwr"
#plt.rcParams['figure.dpi'] = "100"
plt.rcParams['savefig.bbox'] = "tight"


style.use('ggplot')

# Biblioteca para manejar advertencias
import warnings
warnings.filterwarnings('ignore')
```

Importamos el archivo con los datos (en este caso el archivo de E-commerce data)

```python
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
    name=fn
df = pd.read_csv(name, sep=",", encoding="latin1")

#Mostraremos los primeros 5 datos.
df.head()
```

Ahora pasamos a revisar las columnas del archivo, y los tipos de datos utilizados


```python
df.dtypes
```

```python
df['Quantity'].describe()
```

Ahora eliminaremos los datos que no son relevantes en este caso, como seria el stockcode, invoiceNo, y la descripcion

```python
df = df.drop(['StockCode', 'InvoiceNo', 'Description'], axis=1)
df.head(5)
```

Seguido de eso, se eliminaran los datos duplicados, y los nulos

```python
df = df.drop_duplicates()
df = df.dropna()

df.head(5)
```

Una vez hecho la limpieza de datos, se puede comenzar a realizar los desafios:

# **Ticket Promedio:**



```python
# Graficar histograma del total de las compras
plt.figure(figsize=(10, 6))
plt.hist(df['TotalCompra'], bins=30, color='blue', alpha=0.7)
plt.title('Distribución de los Montos de las Compras')
plt.xlabel('Monto de la Compra')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

# **Productos mas vendidos:**

```python
# Graficar un boxplot para comparar la distribución de las compras por país
plt.figure(figsize=(12, 6))
sns.boxplot(x='Country', y='TotalCompra', data=df)
plt.title('Distribución de las Compras por País')
plt.xlabel('País')
plt.ylabel('Monto de la Compra')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# **Diferencia entre paises:**

```python
# Calcular el ticket promedio por país
ticket_promedio_por_pais = df.groupby('Country')['TotalCompra'].mean()

# Calcular las ventas totales por país
ventas_totales_por_pais = df.groupby('Country')['TotalCompra'].sum()

print("Ticket promedio por país:\n", ticket_promedio_por_pais)
print("\nVentas totales por país:\n", ventas_totales_por_pais)
```

# **Comportamiento temporal de las compras**

```python
# Convertir la columna InvoiceDate a tipo datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Resamplear para obtener ventas por mes
ventas_mensuales = df.resample('M', on='InvoiceDate')['TotalCompra'].sum()

# Graficar las ventas mensuales
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
ventas_mensuales.plot()
plt.title("Comportamiento temporal de las compras (mensual)")
plt.xlabel("Fecha")
plt.ylabel("Ventas Totales")
plt.grid(True)
plt.show()
```
